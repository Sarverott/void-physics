# multiscope

aby móc przetwarzać wieloskładnikowe dowody zasadności pokroju matryc tensometrii omniprzestrzennych zależności
trzeba móc załączyć wszystkie znane zasady współzależności równoważnej wartości nośnika i formuły wyrównywania
stanów przestrzennych w przypadku wystąpienia wypadku uwzględnionego jako przyczyn zaistnienia wyrównania kaskadowego

do tego celu będzie potrzebna baza danych zapisów wzorów opisanych matematyk wraz z wagowaniem ich współczynnika pokrycia ze stanem fizycznym - aksiomaty są najwyżej, domysłowe hipotezy najniżej w hierarhii

prócz tego potrzebna będzie baza danych do kolekcjonowania identyfikowalnych reprezentacji typowania wartości wstawianych we wzory

zbieranie przebiegów kalkulacji będzie reprezentacją zdarzeń w tym modelu którego wszystkie składowe procesy, dane i relacje będą wskaźnikiem zarysów istniejącego w rzeczywistości nośnika rzeczywistego
